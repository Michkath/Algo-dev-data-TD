## 4. Couplage

### Rappel
Le couplage représente le **niveau de dépendance entre les classes**.

- **Couplage fort** : les classes dépendent trop les unes des autres
- **Couplage faible** : les classes sont plus indépendantes

### Exercice 4 — Mauvais exemple

```python
class Moteur:
    def demarrer(self):
        print("Moteur ON")

class VoitureCouplageFort:
    def __init__(self):
        self.moteur = Moteur()

    def demarrer(self):
        self.moteur.demarrer()
```

```python
vf = VoitureCouplageFort()
vf.demarrer()
```

## 7. Diamond Problem

## 8. MRO — Method Resolution Order

### Rappel
Le MRO détermine l'ordre dans lequel Python recherche les méthodes.

### Exercice 9

### Interprétation
Python cherche dans l'ordre :
1. `D`
2. `B`
3. `C`
4. `A`
5. `object`

C'est pour cela que `D().action()` appelle d'abord `B.action()`.

## 9. super() et coopération des classes

Dans un contexte d'héritage multiple, `super()` permet aux classes de coopérer
correctement avec le MRO.

```python
class A:
    def action(self):
        print("A")
        super().action()

class Base:
    def action(self):
        print("Fin de chaîne")

class B(A, Base):
    def action(self):
        print("B")
        super().action()

class C(A, Base):
    def action(self):
        print("C")
        super().action()

class D(B, C):
    def action(self):
        print("D")
        super().action()
```

```python
d = D()
d.action()
```

## 10. Composition vs Héritage

### Rappel
- **Héritage** : relation **"est un"**
- **Composition** : relation **"a un"**

### Mauvais exemple
Dire qu'une voiture **est un** moteur n'a pas de sens.

```python
class MoteurSimple:
    def demarrer(self):
        print("Moteur simple démarré")

class MauvaiseVoiture(MoteurSimple):
    pass

mv = MauvaiseVoiture()
mv.demarrer()
```

### Bon exemple
Une voiture **a un** moteur.

```python
class BonneVoiture:
    def __init__(self, moteur):
        self.moteur = moteur

    def demarrer(self):
        print("La voiture transmet l'ordre de démarrage")
        self.moteur.demarrer()

bv = BonneVoiture(MoteurSimple())
bv.demarrer()
```

### Question
Pourquoi la composition est-elle souvent plus flexible que l'héritage ?
