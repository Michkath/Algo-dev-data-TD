## Rappel de cours

### Dijkstra
L’algorithme de **Dijkstra** permet de calculer le **plus court chemin** depuis un sommet source vers tous les autres sommets d’un graphe **pondéré à poids positifs**.

### Conditions d’utilisation
- le graphe peut être orienté ou non orienté ;
- les poids doivent être **positifs ou nuls** ;
- l’algorithme ne fonctionne pas correctement avec des poids négatifs.

### Structure utilisée
- un dictionnaire `distances` pour stocker les distances minimales ;
- une **file de priorité** implémentée avec `heapq`.

### Idée à retenir
> On traite toujours d’abord le sommet dont la distance connue est la plus petite.

## Graphe de travail

Le graphe est représenté sous forme de **liste d’adjacence pondérée** :

- chaque clé est un sommet ;
- chaque valeur est une liste de couples `(voisin, poids)`.

```python
graph = {
    "A": [("B", 4), ("C", 2)],
    "B": [("A", 4), ("C", 1), ("D", 5)],
    "C": [("A", 2), ("B", 1), ("D", 8), ("E", 10)],
    "D": [("B", 5), ("C", 8), ("E", 2), ("F", 6)],
    "E": [("C", 10), ("D", 2), ("F", 3)],
    "F": [("D", 6), ("E", 3)]
}

graph
```

## Code étudié
On part de la version suivante de Dijkstra :

```python
import heapq

def dijkstra(graph, start):
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    pq = [(0, start)]

    while pq:
        current_distance, current_node = heapq.heappop(pq)

        for neighbor, weight in graph[current_node]:
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))

    return distances
```
