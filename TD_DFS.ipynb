## Rappel de cours

### DFS — Depth First Search
Le **DFS** est un **parcours en profondeur**.

### Principe
À partir d’un sommet de départ :
1. on visite le sommet courant ;
2. on visite récursivement un voisin non encore visité ;
3. on continue **le plus loin possible** ;
4. on revient en arrière lorsqu’il n’y a plus de voisin non visité.

### Structure utilisée
- souvent une **pile** implicite via la **récursion** ;
- ou une pile explicite.

### Idée à retenir
> DFS explore une branche jusqu’au bout avant de revenir.

## Graphe de travail
On utilisera d’abord le graphe suivant :

- A est connecté à B et C
- B est connecté à A, D et E
- C est connecté à A et F
- D est connecté à B
- E est connecté à B et F
- F est connecté à C et E

```python
graph = {
    "A": ["B", "C"],
    "B": ["A", "D", "E"],
    "C": ["A", "F"],
    "D": ["B"],
    "E": ["B", "F"],
    "F": ["C", "E"]
}

graph
```

## Code DFS étudié

On part de la version récursive suivante :

```python
def dfs(graph, node, visited=None):
    if visited is None:
        visited = set()

    visited.add(node)

    for neighbor in graph[node]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

    return visited
```
