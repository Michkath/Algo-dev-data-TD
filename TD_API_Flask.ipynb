# **TD — Développement d’une API REST avec Flask et MariaDB**
## **Objectif**

Dans cet exercice, vous allez développer une API REST permettant de gérer des patients dans une base de données MariaDB. <br>

À la fin de cet exercice, vous serez capables de : <br>

+ créer des routes REST avec Flask ; <br>
+ interagir avec une base de données MariaDB ; <br>
+ manipuler des requêtes SQL (SELECT, INSERT, UPDATE, DELETE) ; <br>
+ gérer les erreurs côté serveur ; <br>
+ structurer les réponses JSON. <br>

## **Contexte**

Vous travaillez pour un système hospitalier. <br>
Vous devez développer une API permettant de gérer les patients. <br>

Chaque patient possède les attributs suivants : <br>

- id (entier, auto-incrémenté) <br>
- nom (string) <br>
- age (entier) <br>
- maladie (string) <br>

## **Prérequis**
- Flask installé
- MariaDB installé
- Une base de données hopital
- Une table patients

**Structure SQL attendue :**

```python
CREATE TABLE patients (
    id INT AUTO_INCREMENT PRIMARY KEY,
    nom VARCHAR(100),
    age INT,
    maladie VARCHAR(100)
);
```

## **Travail à réaliser**

Vous devez implémenter les routes suivantes :

### **1. Récupérer tous les patients**
Route :

```python
GET /api/patients
```

**Travail demandé :** <br>
- Se connecter à la base de données
- Exécuter une requête SELECT * FROM patients
- Retourner la liste des patients en JSON
- Gérer les erreurs serveur

### **2. Ajouter un patient**
**Route :**

```python
POST /api/patients
```

**Exemple de requête JSON :**

```python
{
  "nom": "Dupont",
  "age": 45,
  "maladie": "Grippe"
}
```

**Travail demandé :** <br>
- Récupérer les données avec request.get_json()
- Insérer le patient en base (INSERT)
- Retourner :
   - un message de confirmation
   - le patient créé avec son id
- Gérer les erreurs

### **3. Supprimer un patient**
**Route :**

```python
DELETE /api/patients/<id>
```

**Travail demandé :** <br>
- Supprimer le patient correspondant (DELETE)
- Vérifier si le patient existe :
   - si non → retourner 404
- Retourner un message JSON

### **4. Modifier un patient**
**Route :**

```python
PUT /api/patients/<id>
```

**Travail demandé :** <br>
- Récupérer les données envoyées
- Mettre à jour le patient (UPDATE)
- Vérifier si le patient existe :
    - si non → 404
- Retourner le patient mis à jour

**Contraintes techniques** <br>
- Utiliser get_connection() pour accéder à la base
- Toujours :
   - fermer le curseur
   - fermer la connexion (bloc finally)
- Gérer les exceptions avec try / except
