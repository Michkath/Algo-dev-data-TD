# 1. Rappel de cours

## Définition
Le **BFS** (*Breadth First Search*) ou **parcours en largeur** explore un graphe **niveau par niveau** à partir d'un sommet de départ.

## Caractéristiques
- utilise une **file** (*queue*) ;
- visite d'abord les voisins directs ;
- garantit le **plus court chemin en nombre d'arêtes** dans un graphe **non pondéré**.

## Structures utiles
- `visited` : ensemble des sommets déjà visités ;
- `queue` : file des sommets à traiter ;
- `distances` : dictionnaire des distances depuis la source.

# 4. Implémentation correcte de BFS
Voici une version correcte de BFS.

```python
from collections import deque

def bfs(graph, start):
    visited = {start}
    queue = deque([start])
    distances = {start: 0}

    while queue:
        node = queue.popleft()

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                distances[neighbor] = distances[node] + 1
                queue.append(neighbor)

    return distances
```

```python
bfs(graph, "A")
```

# 7. Version commentée pas à pas
Cette version affiche l'évolution de la file et des distances.

```python
from collections import deque

def bfs_trace(graph, start):
    visited = {start}
    queue = deque([start])
    distances = {start: 0}

    print("Départ :", start)
    print("File initiale :", list(queue))
    print("-" * 40)

    while queue:
        node = queue.popleft()
        print("Sommet traité :", node)

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                distances[neighbor] = distances[node] + 1
                queue.append(neighbor)
                print(f"  Ajout de {neighbor} | distance = {distances[neighbor]}")

        print("File actuelle :", list(queue))
        print("Visited :", visited)
        print("Distances :", distances)
        print("-" * 40)

    return distances
```

# 8. Ordre de visite
Modifier BFS pour renvoyer aussi l'ordre de visite.

```python
from collections import deque

def bfs_order(graph, start):
    visited = {start}
    queue = deque([start])
    distances = {start: 0}
    order = []

    while queue:
        node = queue.popleft()
        order.append(node)

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                distances[neighbor] = distances[node] + 1
                queue.append(neighbor)

    return order, distances
```

```python
order, distances = bfs_order(graph, "A")
print("Ordre de visite :", order)
print("Distances :", distances)
```

# 10. BFS et plus court chemin
Le BFS permet aussi de reconstruire un chemin minimal dans un graphe non pondéré.

```python
from collections import deque

def bfs_parents(graph, start):
    visited = {start}
    queue = deque([start])
    parents = {start: None}

    while queue:
        node = queue.popleft()

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                parents[neighbor] = node
                queue.append(neighbor)

    return parents

def build_path(parents, target):
    if target not in parents:
        return None

    path = []
    current = target

    while current is not None:
        path.append(current)
        current = parents[current]

    return path[::-1]
```

```python
parents = bfs_parents(graph, "A")
parents
```

```python
build_path(parents, "F")
```
